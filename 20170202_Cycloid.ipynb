## Cycloid Parametric Function

http://j1w2k3.tistory.com/714

$$x = r ( \theta - sin \theta )$$
$$y = r ( 1 - cos \theta )$$

```julia
using Plotly

times = 1;
theta = [0:0.01:times*2*pi;];
r = 20.0;
x = r * ( theta - sin(theta) );
y = r * ( 1 - cos(theta) );

plot(x,y)
```

```julia
using Plotly
using Interact

times = 1;
theta = [0:0.01:times*2*pi;];
r = 20.0;
x = r * ( theta - sin(theta) );
y = r * ( 1 - cos(theta) );
plot(x,y)


@manipulate for theta_now=0:0.01:times*2*pi
    #x_now = r * ( theta_now - sin(theta_now) );
    y_now = r * ( 1 - cos(theta_now) );
    plot(x, y_now)
end
```
